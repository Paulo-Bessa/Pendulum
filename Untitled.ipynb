```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def henon(x,y,steps=1e4,a=1.4,b=0.3):
    X=[x]
    Y=[y]
    for _ in range(int(steps)):
        x = -X[-1]*X[-1] + b*Y[-1] + a
        y = X[-1]
        X.append(x)
        Y.append(y)
    return np.array([X,Y])
```

```python
ans = henon(0.1,0.1)
plt.plot(ans[0],ans[1],"o",markersize=.2)
```

```python
PE = 2
r1 = 1e-2
r2 = 5e-2
```

```python
pairs = ans.T

def distance(x,y):
    x=np.array(x)
    y=np.array(y)
    assert np.shape(x)==np.shape(y)
    
    ans = np.sqrt(sum(np.square(x-y)))
    return ans

OPI=[]
OPI_index=[]
for i in range(len(pairs)-PE):
    if distance(pairs[i],pairs[i+PE])<=r1:
        OPI.append(pairs[i])
        OPI_index.append(i)
OPI=np.array(OPI)
```

```python
pt=OPI.T
plt.plot(ans[0],ans[1],"o",markersize=.2)
plt.plot(pt[0],pt[1],"o",markersize=4,color="red")
```

```python
OPI_c=[]
for i in OPI_index:
    for j in OPI_index:
        if i!=j:
            ver=1
            for k in range(PE):
                if distance(pairs[i+k],pairs[j+k])>r2:
                    ver=0
            if ver:
                OPI_c.append((pairs[i]+pairs[j])/2)
```
